```python
from PIL import Image
import cv2
import numpy as np
img = cv2.imread("E:/S4/Project/images/1.jpeg")
img1 = cv2.imread("E:/S4/Project/images/2.jpeg")
img2 = cv2.imread("E:/S4/Project/images/3.jpeg")
img3 = cv2.imread("E:/S4/Project/images/4.jpeg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)
cv2.waitKey(0)
cv2.destroyAllWindows()
import matplotlib.pyplot as plt
resized = cv2.resize(gray,(70,50),interpolation=cv2.INTER_AREA)
resized1 = cv2.resize(gray1,(70,50),interpolation=cv2.INTER_AREA)
resized2 = cv2.resize(gray2,(70,50),interpolation=cv2.INTER_AREA)
resized3 = cv2.resize(gray3,(70,50),interpolation=cv2.INTER_AREA)

Title = ["Resize","Resize1","Resize2","Resize3"]
images=[resized,resized1,resized2,resized3]
count=4
for i in range(count):
    plt.subplot(2,2,i+1)
    plt.title(Title[i])
    plt.imshow(images[i])
    plt.show()
```
